```python
import pandas as pd
import numpy as np
df = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_161231.txt')
```

```python
df.info()
```

```python
df.columns = df.columns.str.strip()

df.columns
```

```python
df['TIME'].unique()
```

```python
df_sorted = df.sort_values(by=['C/A', 'UNIT', 'SCP', 'STATION', 'DATE', 'TIME']) 
df_sorted
```

```python
df_sorted['DESC'].unique()
```

```python
df_sorted['entry'] = df_sorted.groupby(['C/A', 'UNIT', 'SCP', 'STATION'])['ENTRIES'].transform(lambda x: x.sum())
```

```python
df_sorted['entry'] = df_sorted.groupby(['C/A', 'UNIT', 'SCP', 'STATION'])['ENTRIES'].diff()
df_sorted['exit'] = df_sorted.groupby(['C/A', 'UNIT', 'SCP', 'STATION'])['EXITS'].diff()
upper_bound_entry = df_sorted['entry'].median() + df_sorted['entry'].std() * 4
upper_bound_exit = df_sorted['exit'].median() + df_sorted['exit'].std() * 4

mask = (df_sorted.exit >= 0) & (df_sorted.entry >= 0) & (df_sorted.entry < upper_bound_entry) & (df_sorted.exit < upper_bound_exit)
df_filter = df_sorted[mask]
df_filter.head(10)

# for k, v in df_group:
#     v['ENTRIES'] = v['ENTRIES'].diff()
#     break
# df_group.head(100)
```

```python
df_final = df_filter.groupby(['STATION','DATE', 'TIME', 'LINENAME'], as_index=False)['entry', 'exit'].sum()
df_final.head(10)
```

```python
df_final['DATE'] = pd.to_datetime(df_final['DATE'], format='%m/%d/%Y')
df_final['weekday'] = df_final['DATE'].dt.dayofweek
df_final.head(10)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image

import seaborn as sns

%matplotlib inline
df_top = df_final.groupby(['STATION', 'LINENAME'], as_index=False)['entry', 'exit'].mean()
df_top = df_top.sort_values(by=['entry'], ascending=[False]) 
df_top.head(20)
```
